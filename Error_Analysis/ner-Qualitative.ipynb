```python
from nervaluate import Evaluator
from utils import get_gold_data, get_pred_data, find_missed_entities, bio_to_entities, find_spurious_entities
from collections import Counter
```

```python
lang = "eu"
tokens, true = get_gold_data(lang)
```

```python
for model in ['mbert', 'xlmroberta', 'biobert', f'mbert_multilingual_{lang}', f'xlmroberta_multilingual_{lang}', f'biobert_multilingual_{lang}']:
    print(model)
    pred = get_pred_data(lang, model)
    missed_entities = find_missed_entities(tokens, true, pred)
    # print([x for x in Counter(missed_entities['RML']).most_common() if x[1] > 1])
    print(Counter(missed_entities['TST']).most_common())
    # print(Counter(missed_entities['RML']).most_common())
    print()
```

```python
for model in ['mbert', 'xlmroberta', 'biobert', f'mbert_multilingual_{lang}', f'xlmroberta_multilingual_{lang}', f'biobert_multilingual_{lang}']:
    print(model)
    pred = get_pred_data(lang, model)
    missed_entities = find_spurious_entities(tokens, true, pred)
    # print([x for x in Counter(missed_entities['TST']).most_common() if x[1] > 1])
    print(Counter(missed_entities['TST']).most_common(5))
    # print(Counter(missed_entities['RML']).most_common(5))
    print()
```

```python
tst_entities = []
rml_entities = []
for sent_tokens, sent_labels in zip(tokens, true):
    for entity in bio_to_entities(sent_tokens, sent_labels):
        if entity[0] == 'TST':
            tst_entities.append(entity[1])
        else:
            rml_entities.append(entity[1])
```
