```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from utils import get_error_type_df


lang = "es"

# Create DataFrame
df = get_error_type_df(lang)

```

```python
# Plotting
plt.figure(figsize=(15, 6))
# sns.barplot(data=df, x='Model', y='Count', hue='Error Type')
sns.barplot(data=df, x='Error Type', y='Count', hue='Model')
plt.title('Comparison of Error Types Across Models')
# plt.xlabel('Model')
plt.xlabel('Error Type')
plt.ylabel('Count of Errors')
# plt.legend(title='Error Type')
plt.legend(title='Model')
plt.savefig(f"plots/{lang}_re_error_counts.pdf", format='pdf', bbox_inches='tight')
# plt.show()
```
